#1) What is hypothesis testing in statistics ?
  - Hypothesis testing in statistivs is a method used to make decisions about population parameters based on sample data.



#2) What is null hypothesis, and how does it differ from the alternative hypothesis ?
 - A null hypothesis is a foundational concept in statistics that assumes there is no real relationship or effect in the data being analyzed, and that any variations or trends are simply the result of random fluctution rather than a true underlying cause. It differs from the alternative hypothesis as null hypothesis assumes no effect or no difference where alternative hypothesis assumes there is an effect or a difference.



#3) Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
 - The significance level in hypothesis testing, usually denoted by alpha, is the threshold for deciding whether to reject the null hypothesis. The significance level plays a critical role in deciding the outcome of a hypothesis test by setting the cutoff point for rejecting the null hypothesis.



#4) What are Type I and Type II errors ? Give example of each.
  - Type I error is the rejection of null hypothesis when it is actually true. Example : An innocent person is convicted.
   - Type II error is the failure to reject the null hypothesis that is actually false. Example : A guilty person may not be convicted.



#5) What is the difference between a Z-test and a T-test ? Explain when to use each.
  - Z-test is used when sample size is large (n>30) and population standard deviation is known and uses z-distribution while T-test is used when sample is small (n<= 30) and population standard deviation is unknown and uses t-distribution.

```python
#9) Write a python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.


import numpy as np
import matplotlib.pyplot as plt

def calculate_zscore_and_plot(data):

    data=np.array(data)
    mean=np.mean(data)
    std=np.std(data,ddof=1)

    z_score=(data-mean)/std

    plt.hist(z_score, bins=20, color="blue", edgecolor="black")
    plt.xlabel("z-score")
    plt.ylabel("density")
    plt.axvline(0,color="red", linestyle="--", label="mean (z=0)")
    plt.legend()
    plt.show()

    return z_score

sample_data= [49.1,50.2,51.0,48.7,50.5,49.8,50.3,50.7,50.2,49.6,50.1,49.9,50.8,50.4,48.9,50.6,50.0,49.7,50.2,49.5]

z_score= calculate_zscore_and_plot(sample_data)
print("first 5 z-scores:", z_score[:5].round(2))

```
